```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
cell_phone_data = {
    'Year': [2001, 2006, 2011, 2016, 2021],
    'Cell Phones per Person': [0.15, 0.40, 0.75, 1.05, 1.30] 
}

df_cell_phones = pd.DataFrame(cell_phone_data)
df_cell_phones.set_index('Year', inplace=True)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_cell_phones, marker='o', linewidth=2, color='skyblue')
plt.title('Rise in Cell Phone Usage Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14) 
plt.ylabel('Cell Phones per Person', fontsize=14)
plt.xticks(df_cell_phones.index, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)  
sns.despine()
for year, value in df_cell_phones['Cell Phones per Person'].items():
    plt.annotate(f'{value:.2f}', (year, value), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
population_data = {
    'Year': [2001, 2006, 2011, 2016, 2021],
    'Population (Billions)': [6.2, 6.6, 7.0, 7.4, 7.8]
}

df_population = pd.DataFrame(population_data)
df_population.set_index('Year', inplace=True)


plt.figure(figsize=(10, 6))
sns.lineplot(data=df_cell_phones, marker='o', linewidth=2, color='skyblue', label='Cell Phones per Person')
sns.lineplot(data=df_population, marker='x', linewidth=2, color='coral', label='World Population (Billions)') 

plt.title('Cell Phone Usage vs. Population Growth', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value', fontsize=14) 
plt.xticks(df_cell_phones.index, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
sns.despine()
plt.legend(fontsize=12)

plt.show()
```
