```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = {
    'Year': [1995, 2000, 2005, 2010, 2015],
    'Happiness Score (Afghanistan)': [47.6, 38.3, 37.8, 35.7, 31.3],
    'Happiness Score (India)': [71, 71.2, 72.2, 66.2, 65.4],
    'Happiness Score (Albania)': [64.4, 67.8, 64.7, 65.8, 66.7],
    'Happiness Score (USA)': [74.5, 74.1, 72, 73.6, 72.9],
    'Happiness Score (GERMANY)': [32.3, 38.7, 31.9, 34.8, 33.5],
    'Happiness Score (Australia)': [73, 74.7, 74, 75, 69.5],
    'Happiness Score (Belgium)': [68.5, 71.1, 69.3, 71, 68.5]
}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)
plt.title('Happiness Score Over Time by Country')
plt.xlabel('Happiness Score (Years)')
plt.ylabel('Year')
plt.legend(title='Country')
plt.show()
```
