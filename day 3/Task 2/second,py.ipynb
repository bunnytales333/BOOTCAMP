```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = {
    'Year': [2001, 2006, 2011, 2016, 2021],
    'Average Income (USA)': [77000, 77500, 79500, 80000, 79000],
    'Average Income (INDIA)': [60000, 50000, 81000, 75000, 82500],
    'Average Income (GERMANY)': [70000, 72500, 80000, 66000, 81200]
}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)
plt.title('Average Income Over Time by Country')
plt.xlabel('Average Income (Years)')
plt.ylabel('Year')
plt.legend(title='Country')
plt.show()
```
